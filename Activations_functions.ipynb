## Activation Functions

Activation functions are crucial components of neural networks that introduce non-linearity into the model, allowing it to learn complex patterns. Without them, a neural network would simply be a linear regression model.

Here, we'll explore some common activation functions and their implementations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a range of input values for plotting
x = np.linspace(-5, 5, 100)
```

### 1. Sigmoid (Logistic) Activation Function

**Formula:** $f(x) = \frac{1}{1 + e^{-x}}$

**Characteristics:**
*   Output values are in the range (0, 1).
*   Commonly used in the output layer of binary classification models as it can be interpreted as a probability.
*   Suffers from the vanishing gradient problem for very large or very small inputs.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Apply Sigmoid
y_sigmoid = sigmoid(x)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(x, y_sigmoid)
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.show()
```

### 2. Tanh (Hyperbolic Tangent) Activation Function

**Formula:** $f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$

**Characteristics:**
*   Output values are in the range (-1, 1).
*   Similar to Sigmoid but outputs are zero-centered, which can make training easier for subsequent layers.
*   Also suffers from the vanishing gradient problem.

```python
def tanh(x):
    return np.tanh(x)

# Apply Tanh
y_tanh = tanh(x)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(x, y_tanh)
plt.title('Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.grid(True)
plt.show()
```

### 3. ReLU (Rectified Linear Unit) Activation Function

**Formula:** $f(x) = \max(0, x)$

**Characteristics:**
*   Output is 0 for negative inputs and $x$ for positive inputs.
*   Widely popular due to its computational efficiency and ability to mitigate the vanishing gradient problem compared to Sigmoid and Tanh.
*   Can suffer from the "dying ReLU" problem where neurons can become inactive for negative inputs.

```python
def relu(x):
    return np.maximum(0, x)

# Apply ReLU
y_relu = relu(x)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(x, y_relu)
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('relu(x)')
plt.grid(True)
plt.show()
```

### 4. Leaky ReLU Activation Function

**Formula:** $f(x) = \begin{cases} x & \text{if } x > 0 \\ \alpha x & \text{if } x \le 0 \end{cases}$

**Characteristics:**
*   A variant of ReLU that allows a small, non-zero gradient for negative inputs (controlled by $\alpha$, typically 0.01).
*   Addresses the "dying ReLU" problem by preventing neurons from being completely inactive.
*   Computationally efficient.

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, x * alpha)

# Apply Leaky ReLU
y_leaky_relu = leaky_relu(x)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(x, y_leaky_relu)
plt.title('Leaky ReLU Activation Function (alpha=0.01)')
plt.xlabel('x')
plt.ylabel('leaky_relu(x)')
plt.grid(True)
plt.show()
```

### 5. Softmax Activation Function

**Formula:** $f(x_i) = \frac{e^{x_i}}{\sum_{j=1}^{K} e^{x_j}}$ (for each element $x_i$ in an input vector $\mathbf{x}$ of size $K$)

**Characteristics:**
*   Transforms a vector of arbitrary real values into a probability distribution.
*   Output values are in the range (0, 1) and sum up to 1.
*   Most commonly used in the output layer of multi-class classification models.

```python
def softmax(x):
    # Subtract max for numerical stability
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

# Softmax is typically applied to an array of scores (logits)
# Let's create an example input array
input_scores = np.array([1.0, 2.0, 3.0, 4.0])
y_softmax = softmax(input_scores)

print(f"Input Scores: {input_scores}")
print(f"Softmax Output: {y_softmax}")
print(f"Sum of Softmax Output: {np.sum(y_softmax)}")

# Plotting (for conceptual understanding, not the function itself)
plt.figure(figsize=(8, 4))
plt.bar(range(len(input_scores)), y_softmax)
plt.title('Softmax Activation Function Output (Example)')
plt.xlabel('Class Index')
plt.ylabel('Probability')
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```

This overview covers the most frequently used activation functions. The choice of activation function depends on the specific problem, the network architecture, and desired output characteristics.
